#Building the model

```python
x=med_ins.drop('charges',axis=1)
y=med_ins['charges']
x.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
med_ins
```

##linear regression

```python
lr=LinearRegression().fit(x_train,y_train)
pred=lr.predict(x_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)

print("the MAE with linearRegression: "+str(s))
print("the MSE with LinearRegression: "+str(s1))
print("the r2_score: "+str(s2))
```

##Decision Tree Regressor

```python
dtr=DecisionTreeRegressor()          
dtr=dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)


print("the MAE is: "+str(s))
print('the MsE is: '+str(s1))
print('the r2 is: '+str(s2))
```

##random tree regressor

```python
rdm=RandomForestRegressor()
rdm=rdm.fit(x_train,y_train)
pred=rdm.predict(x_test)
s=mean_absolute_error(y_test,pred)
s1=mean_squared_error(y_test,pred)
s2=r2_score(y_test,pred)


print('the MAE value'+str(s))
print('the MAE value'+str(s1))
print('the MAE value'+str(s2))
```
